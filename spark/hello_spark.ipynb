```python
from pyspark.sql import SparkSession
```

```python
spark = (
    SparkSession.builder
    .appName("Ecommerce Bronze Reader")
    .master("spark://spark-master:7077")
    .getOrCreate()
)

df = spark.read.csv(
    "s3a://bronze-layer/olist_order_items_dataset.csv",
    header=True
)
df.cache()
df.count() # Ép spark cache vì là đây là action
df.printSchema()
df.show(5)
```
